```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')

data = 'C:/Users/peter/Downloads/National_Child_Abuse_and_Neglect_Data_System__NCANDS__Child_File__Findings_from_the_National_Child_Abuse_and_Neglect_Data_System__NCANDS__-_Data_Tables__updated_with_FY_2015_.csv'
df = pd.read_csv(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
tableDict = {}

for i in df.Measure.unique():
    tableDict[i] = df[df['Measure'] == i]
```

```python
m_keys = {}

for i in range(len(list(tableDict.keys()))):
    m_keys[i] = list(tableDict.keys())[i]
```

```python
m_keys
```

```python
df['Characteristic Label'].unique()
```

```python
tableDict[m_keys[13]].groupby(
    'Characteristic Label').sum().sort_values('Value', ascending = False).head(3).plot.bar(y='Value', 
                                           figsize = (10,5), 
                                           ylabel = 'Total Cases',
                                          xlabel = 'Perpetrators',
                                          title = "Top 3 Perpetrators")
```

```python
tableDict[m_keys[13]].groupby('State').sum().sort_values('Value', ascending=False)[1:].head(5).plot.bar(y='Value',
                                                                figsize=(10,5),
                                                                ylabel="Total Child Fatalities",
                                                                xlabel="State",
                                                                title="5 Year Child Abuse Fatality Data"
                                                                )
```

```python
top_states = ['California', 'New York', 'Texas', 'Florida', 'Michigan']
```

```python
df = tableDict[m_keys[13]]

for state in top_states:
    df[df['State'] == state].groupby(
    'Characteristic Label').sum().sort_values('Value', ascending = False).head(3).plot.bar(y='Value', 
                                           figsize = (10,5), 
                                           ylabel = 'Total Cases',
                                          xlabel = 'Perpetrators',
                                          title = "Top 3 Perpetrators for {}".format(state))
    
```
